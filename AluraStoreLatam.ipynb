````python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
````

# 2. Ventas por categoría

````python
#Por Monto de Facturacion de productos Vendidos por Categoria en cada Tienda

#Genera lista de todas las categorias vendidas y los montos
list_ventas_por_categoria_tienda1=list(tienda1['Categoría del Producto'])
list_ventas_por_categoria_tienda2=list(tienda2['Categoría del Producto'])
list_ventas_por_categoria_tienda3=list(tienda3['Categoría del Producto'])
list_ventas_por_categoria_tienda4=list(tienda4['Categoría del Producto'])
list_monto_por_categoria_tienda1=list(tienda1['Precio'])
list_monto_por_categoria_tienda2=list(tienda2['Precio'])
list_monto_por_categoria_tienda3=list(tienda3['Precio'])
list_monto_por_categoria_tienda4=list(tienda4['Precio'])
def CalcularMonto(tienda,lista_ventas,lista_monto):
  ventas_por_categoria = {}
  for i in range(len(lista_ventas)):
    categoria = lista_ventas[i]
    monto = lista_monto[i]

    if categoria in ventas_por_categoria:
      ventas_por_categoria[categoria] += monto
    else:
      ventas_por_categoria[categoria] = monto
  ventas_ordenadas= sorted(ventas_por_categoria.items(), key=lambda x: x[1], reverse=True)


  #Mostrar Resultados
  print("--------")
  print(f"{tienda}:")
  print("--------")
  for categoria, total in ventas_ordenadas:
    print(f"{categoria}: ${total}")

  # Separar categorías y montos
  categorias = [item[0] for item in ventas_ordenadas]
  montos = [item[1] for item in ventas_ordenadas]

  # Crear gráfico
  plt.figure(figsize=(8, 6))
  barras=plt.bar(categorias, montos, color='skyblue')
  plt.xlabel('Categorías')
  plt.ylabel('Monto total vendido $')
  plt.title(f'Ventas por Categoría de la {tienda}')
  plt.xticks(rotation=45)
  # Agregar valores encima de cada barra
  for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, altura + 5000000,  # ajustar el desplazamiento vertical si es necesario
             f'{int(altura):,}', ha='center', va='bottom', fontsize=10)
  plt.tight_layout()

  plt.show()


CalcularMonto("Tienda1",list_ventas_por_categoria_tienda1,list_monto_por_categoria_tienda1)

````

# 4. Productos más y menos vendidos

````python
#Por Monto de Facturacion :productos mas Vendidos y menos vendidos en cada Tienda

#Genera lista de todas las categorias vendidas y los montos
list_ventas_por_producto_tienda1=list(tienda1['Producto'])
list_ventas_por_producto_tienda2=list(tienda2['Producto'])
list_ventas_por_producto_tienda3=list(tienda3['Producto'])
list_ventas_por_producto_tienda4=list(tienda4['Producto'])
list_monto_por_producto_tienda1=list(tienda1['Precio'])
list_monto_por_producto_tienda2=list(tienda2['Precio'])
list_monto_por_producto_tienda3=list(tienda3['Precio'])
list_monto_por_producto_tienda4=list(tienda4['Precio'])
def CalcularMonto(tienda,lista_ventas,lista_monto):
  ventas_por_categoria = {}
  for i in range(len(lista_ventas)):
    categoria = lista_ventas[i]
    monto = lista_monto[i]

    if categoria in ventas_por_categoria:
      ventas_por_categoria[categoria] += monto
    else:
      ventas_por_categoria[categoria] = monto
  ventas_ordenadas= sorted(ventas_por_categoria.items(), key=lambda x: x[1], reverse=True)
  ventas_ordenadas_mas=ventas_ordenadas[0:3]
  ventas_ordenadas_menos=ventas_ordenadas[-3:]

  #Mostrar Resultados
  print("--------")
  print(f"{tienda}:")
  print("--------")
  for categoria, total in ventas_ordenadas_mas+ventas_ordenadas_menos:
    print(f"{categoria}: ${total}")

  # Separar categorías y montos
  categorias_mas = [item[0] for item in ventas_ordenadas_mas]
  montos_mas = [item[1] for item in ventas_ordenadas_mas]

  categorias_menos = [item[0] for item in ventas_ordenadas_menos]
  montos_menos = [item[1] for item in ventas_ordenadas_menos]
  # Crear gráfico
  plt.figure(figsize=(8, 6))
  barras_mas=plt.bar(categorias_mas, montos_mas, color='blue')
  barras_menos=plt.bar(categorias_menos, montos_menos, color='skyblue')

  plt.xlabel('Producto')
  plt.ylabel('Monto total vendido $')
  plt.title(f'Ventas por Producto de la {tienda}')
  plt.xticks(rotation=45)
  # Agregar valores encima de cada barra
  for barra in barras_mas:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, altura + 5000000,  # ajustar el desplazamiento vertical si es necesario
             f'{int(altura):,}', ha='center', va='bottom', fontsize=10)
  for barra in barras_menos:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, altura + 5000000,  # ajustar el desplazamiento vertical si es necesario
             f'{int(altura):,}', ha='center', va='bottom', fontsize=10)
  plt.tight_layout()

  plt.show()


CalcularMonto("Tienda1",list_ventas_por_producto_tienda1,list_monto_por_producto_tienda1)
````

# 5. Envío promedio por tienda

````python
# Genera lista con los costos de envío de cada tienda
lista_envios_tienda1 = list(tienda1['Costo de envío'])
lista_envios_tienda2 = list(tienda2['Costo de envío'])
lista_envios_tienda3 = list(tienda3['Costo de envío'])
lista_envios_tienda4 = list(tienda4['Costo de envío'])

# Suma los costos de envío de cada tienda
suma_envios_tienda1 = sum(lista_envios_tienda1)
suma_envios_tienda2 = sum(lista_envios_tienda2)
suma_envios_tienda3 = sum(lista_envios_tienda3)
suma_envios_tienda4 = sum(lista_envios_tienda4)

# Cuenta la cantidad de elementos de cada lista
cantidad_envios_tienda1 = len(lista_envios_tienda1)
cantidad_envios_tienda2 = len(lista_envios_tienda2)
cantidad_envios_tienda3 = len(lista_envios_tienda3)
cantidad_envios_tienda4 = len(lista_envios_tienda4)

# Realiza promedio de los costos de envío de cada tienda
promedio_envios_tienda1 = suma_envios_tienda1 / cantidad_envios_tienda1
promedio_envios_tienda2 = suma_envios_tienda2 / cantidad_envios_tienda2
promedio_envios_tienda3 = suma_envios_tienda3 / cantidad_envios_tienda3
promedio_envios_tienda4 = suma_envios_tienda4 / cantidad_envios_tienda4

# Imprime el promedio de envío por tienda
print(f"El costo de envío promedio de la tienda 1 es de {promedio_envios_tienda1:,.2f}")
print(f"El costo de envío promedio de la tienda 2 es de {promedio_envios_tienda2:,.2f}")
print(f"El costo de envío promedio de la tienda 3 es de {promedio_envios_tienda3:,.2f}")
print(f"El costo de envío promedio de la tienda 4 es de {promedio_envios_tienda4:,.2f}")

````

````python
import matplotlib.pyplot as plt

# Nombres de las tiendas
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Promedios ya calculados
promedios = [
    promedio_envios_tienda1,
    promedio_envios_tienda2,
    promedio_envios_tienda3,
    promedio_envios_tienda4
]

# Crear gráfico de barras
plt.figure(figsize=(8, 5))
barras = plt.bar(tiendas, promedios, color='skyblue')

# Título y etiquetas
plt.title('Costo de envío promedio por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Costo promedio de envío ($)')
plt.ylim(0, max(promedios) * 1.1)  # Un poco más alto para que entren los valores

# Mostrar el valor numérico encima de cada barra
for barra, valor in zip(barras, promedios):
    plt.text(barra.get_x() + barra.get_width()/2, barra.get_height(),
             f"${valor:,.2f}", ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
````
